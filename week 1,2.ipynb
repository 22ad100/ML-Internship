```python
print("*********************************************************Data Loading**************************************************************")
import pandas as pd
a= pd.read_csv("nba.csv")
print(a)
print(a["TARGET_5Yrs"])
a.isna()
```

```python
a.fillna(value=0, method=None, inplace=True)
```

```python
from sklearn import preprocessing
b=preprocessing.LabelEncoder()
a["Name"]=b.fit_transform(a["Name"])
```

```python
from matplotlib import pyplot as plt
plt.plot(a["Name"],a["TARGET_5Yrs"])
```

```python
x=a.iloc[:, :-1].values
print(x)
```

```python
y=a["TARGET_5Yrs"]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30, shuffle=True)
print(x_train)
print(x_test)
```

```python
print("*****************************************************LOGISTIC REGRESSION***********************************************************")
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
s=LogisticRegression()
s.fit(x_train, y_train)
predictions=s.predict(x_test)
print("accuracy",accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
print("*************************************DECISION TREE CLASSIFIER*******************************************")

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
s=DecisionTreeClassifier()
s.fit(x_train, y_train)
predictions=s.predict(x_test)
print("accuracy",accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
print("**********************************************RANDOM FOREST CLASSIFIER**************************************************")

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
s=RandomForestClassifier()
s.fit(x_train, y_train)
predictions=s.predict(x_test)
print("accuracy",accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
print("****************************************NAIVE BAYES*************************************************")

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
s=GaussianNB()
s.fit(x_train, y_train)
predictions=s.predict(x_test)
print("accuracy",accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
print("******************************************SVM*************************************************")

from sklearn import svm
from sklearn.metrics import accuracy_score
s=svm.SVC()
s.fit(x_train, y_train)
predictions=s.predict(x_test)
print("accuracy",accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
print("******************************************KNN*************************************************")
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
s=KNeighborsClassifier()
s.fit(x_train, y_train)
predictions=s.predict(x_test)
print("accuracy",accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
